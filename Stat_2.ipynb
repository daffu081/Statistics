```python
import pandas as pd
import scipy.stats as stats
```

```python
data = pd.read_excel("/content/drive/MyDrive/data science/Concrete_Data.xls")
```

```python
stats_summary = {}
```

```python
for column in data.columns:
    feature_data = data[column]
    stats_summary[column] = {
        'Mean': feature_data.mean(),
        'Median': feature_data.median(),
        'Mode': feature_data.mode()[0] if not feature_data.mode().empty else None,
        'Range': feature_data.max() - feature_data.min(),
        'Variance': feature_data.var(),
        'Standard Deviation': feature_data.std(),
        'Interquartile Range': feature_data.quantile(0.75) - feature_data.quantile(0.25),
        'Skewness': feature_data.skew(),
        'Kurtosis': feature_data.kurtosis()
    }

# Display the summary
stats_summary
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Adjust plot size for better readability
plt.figure(figsize=(18, 36))

# Loop through each feature to create histograms, boxplots, and scatter plots
for i, col in enumerate(data.columns):
    plt.subplot(9, 3, i*3 + 1)
    sns.histplot(data[col], kde=True, color="skyblue")
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)

    plt.subplot(9, 3, i*3 + 2)
    sns.boxplot(x=data[col], color="lightgreen")
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

    if col != 'Concrete compressive strength(MPa, megapascals) ':
        plt.subplot(9, 3, i*3 + 3)
        sns.scatterplot(x=data[col], y=data['Concrete compressive strength(MPa, megapascals) '])
        plt.title(f'{col} vs. Concrete Strength')
        plt.xlabel(col)
        plt.ylabel("Concrete Strength (MPa)")

plt.tight_layout()
plt.show()
```
