```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/data science/shipping data.csv')
print(df.head(10))
```

```python
# Extract the 'Weight in gms' column
weights = df['Weight_in_gms']

# Calculate the range of the data
range_weights = weights.max() - weights.min()

# Determine the number of classes (you can adjust this based on your preference)
num_classes = 5

# Calculate the class width
class_width = range_weights / num_classes

# Create class intervals
intervals = []
lower_bound = weights.min()
for i in range(num_classes):
    upper_bound = lower_bound + class_width
    intervals.append((lower_bound, upper_bound))
    lower_bound = upper_bound

# Create the frequency table
frequency_table = {}
for interval in intervals:
    frequency_table[str(interval)] = 0

for weight in weights:
    for interval in intervals:
        if interval[0] <= weight < interval[1]:
            frequency_table[str(interval)] += 1

# Print the grouped frequency distribution table
print("Grouped Frequency Distribution Table for Weight in gms:")
print("-----------------------------------------------------")
print("Class Interval | Frequency")
print("-----------------------------------------------------")
for interval, frequency in frequency_table.items():
    print(f"{interval} | {frequency}")
```

```python
# Contingency table between Warehouse_block and Mode_of_Shipment
contingency_table = pd.crosstab(df['Warehouse_block'], df['Mode_of_Shipment'])

print(contingency_table)
```

```python
# Bar chart of “Warehouse_block” and “Product_importance”

import matplotlib.pyplot as plt

# Create a bar chart for 'Warehouse_block' and 'Product_importance'
df.groupby(['Warehouse_block', 'Product_importance']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Warehouse Block and Product Importance Distribution')
plt.xlabel('Warehouse Block')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Cost_of_the_Product
plt.hist(df['Cost_of_the_Product'], bins=10, edgecolor='black')
plt.title('Histogram of Cost of the Product')
plt.xlabel('Cost of the Product')
plt.ylabel('Frequency')
plt.show()

# Histogram for Weight_in_gms
plt.hist(df['Weight_in_gms'], bins=10, edgecolor='black')
plt.title('Histogram of Weight in gms')
plt.xlabel('Weight in gms')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot of “Cost_of_the_Product” and “Weight in gms” individually. (2301-2400 datapoints)


df_subset = df.loc[2300:2400]  # rows from 2301 to 2400

# Scatter plot for 'Cost_of_the_Product' and 'Weight_in_gms'
plt.figure(figsize=(8, 6))  # Optional: Adjust figure size
plt.scatter(df_subset['Cost_of_the_Product'], df_subset['Weight_in_gms'])
plt.xlabel('Cost_of_the_Product')
plt.ylabel('Weight_in_gms')
plt.title('Scatter Plot of Cost vs. Weight')
plt.show()
```
